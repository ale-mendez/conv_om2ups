# Operate over new dataframe with data of interest

```python
# remove useless columns:
dat1.drop(columns=['E1-DATA','S', 'ALPHA(POL)'],inplace=True)
# take absolute value of gf and add its logarithm as new column
dat1['G*F']=abs(dat1['G*F'])
dat1['A(EK)*SEC']=abs(dat1['A(EK)*SEC'])
dat1['log(GF)']=np.log(dat1['G*F'])
# compute the relative error of GF in length and velocity gauge in new column:
dat1['ER% G*F']=(abs(dat1.loc[:]['G*F'])-abs(dat1.loc[:]['GF(VEL)']))/abs(dat1.loc[:]['G*F'])*100
```

```python
kdx_dat1=dat1.loc[:]['K'].tolist()
tcol=[]
jcol=[]
for i in kdx_dat1:
    dummy=LEV.loc[LEV.iloc[:]['K']==i][['K','T','J']]
    dumm=dummy.values[0]
    tcol.append(int(dumm[1]))
    jcol.append(int(dumm[2]))
dat1['T']=tcol
dat1['J']=jcol
```

```python
kpdx_dat1=dat1.loc[:]['KP'].tolist():
tpcol=[]
jpcol=[]
for i in kpdx_dat1:
    dummy=LEV.loc[LEV.iloc[:]['K']==i][['K','T','J']]
    dumm=dummy.values[0]
    tpcol.append(int(dumm[1]))
    jpcol.append(int(dumm[2]))
dat1['TP']=tpcol
dat1['JP']=jpcol
```

```python
as2db={44:[27,1],                     # 3s5g-1G
       45:[26,1],46:[26,3],47:[26,2], # 3s5g-3G
       69:[40,1],                     # 3s6h-1H
       70:[39,1],71:[39,3],72:[39,2], # 3s6h-3H
       65:[38,1],                     # 3s6g-1G
       66:[37,3],67:[37,1],68:[37,2], # 3s6g-3G
       91:[52,1],                     # 3s7h-1H
       92:[51,1],93:[51,3],94:[51,2], # 3s7h-3H
       95:[54,1],                     # 3s7i-1I
       96:[53,1],97:[53,3],98:[53,2], # 3s7i-3I
       } 
```

```python
database_updated=database.copy()
print('        database     AS')
for i in range(ndb):
    print(i)
    dumm=database.iloc[i:i+1][:]
    lwlv=dumm.iloc[0]['LowerLevel']
    lwsub=dumm.iloc[0]['LowerSublevel']
    uplv=dumm.iloc[0]['UpperLevel']
    upsub=dumm.iloc[0]['UpperSublevel']
    for (auto, dbase) in as2db.items():
        if (lwlv==dbase[0]) & (lwsub==dbase[1]):
            lw_as=auto
            print('lower=  ',int(lwlv),",",int(lwsub)," == ",auto)
        if (uplv==dbase[0]) & (upsub==dbase[1]):
            up_as=auto
            print('upper=  ',int(uplv),",",int(upsub)," == ",auto)
    dum2=dat1.loc[:][(dat1.loc[:]['K']==up_as)&(dat1.loc[:]['KP']==lw_as)]
    as_gf=dum2.loc[:]['G*F'].tolist()
    as_A=dum2.loc[:]['A(EK)*SEC'].tolist()
    database_updated.at[i,'gf']=as_gf[0]
    database_updated.at[i,'A']=as_A[0]
    print(' ')
```

```python
database_updated.to_csv(r'LineFine-updated.dat', header=None, index=None, sep='\t', mode='a')
```
