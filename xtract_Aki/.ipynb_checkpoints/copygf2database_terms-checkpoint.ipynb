# Operate over new dataframe with data of interest

```python
# remove useless columns:
dat1.drop(columns=['E1-DATA','S', 'ALPHA(POL)'],inplace=True)
# take absolute value of gf and add its logarithm as new column
dat1['G*F']=abs(dat1['G*F'])
dat1['A(EK)*SEC']=abs(dat1['A(EK)*SEC'])
dat1['log(GF)']=np.log(dat1['G*F'])
# compute the relative error of GF in length and velocity gauge in new column:
dat1['ER% G*F']=(abs(dat1.loc[:]['G*F'])-abs(dat1.loc[:]['GF(VEL)']))/abs(dat1.loc[:]['G*F'])*100
```

# Input of database transition data to be updated

```python
database=pd.read_csv("Line-um.dat",sep='\t',header='infer')
ndb=len(database)
```

```python
database
```

```python
# en DataBase:    en AS:
#
# 3s5g-3G:
#  LwLev=26       T=15   I=26
#
# 3s5g-1G:
#  LwLv=27        T=20   I=27
#
# ---------------------------------
#
# 3s6h-3H:
#  LwLv=39        T=7   I=39
#
# 3s6h-1H:
#  LwLv=40        T=11   I=40
#
# 3s6g-3G:
#  LwLev=37       T=16   I=38
#
# 3s6g-1G:
#  LwLv=38        T=21   I=37
#
# ---------------------------------
#
# 3s7h-3H:
#  LwLv=51        T=8   I=52
#
# 3s7h-1H:
#  LwLv=52        T=12   I=51
#       
# 3s7i-3I:
#  LwLv=53        T=3   I=53
#
# 3s7i-1I:
#  LwLv=54        T=5   I=54
#       
```

```python
as2db={27:27,                     # 3s5g-1G
       26:26,                     # 3s5g-3G
       40:40,                     # 3s6h-1H
       39:39,                     # 3s6h-3H
       37:38,                     # 3s6g-1G
       38:37,                     # 3s6g-3G
       51:52,                     # 3s7h-1H
       52:51,                     # 3s7h-3H
       54:54,                     # 3s7i-1I
       53:53,                     # 3s7i-3I
       } 
```

```python
database_updated=database.copy()
print('        database     AS')
for i in range(ndb):
    print(i)
    dumm=database.iloc[i:i+1][:]
    lwlv=dumm.iloc[0]['LowerLevel']
    uplv=dumm.iloc[0]['UpperLevel']
    print(lwlv,"=>",uplv)
    for (auto, dbase) in as2db.items():
        if lwlv==dbase:
            lw_as=auto
            print('lower=  ',int(lwlv)," == ",auto)
        if uplv==dbase:
            up_as=auto
            print('upper=  ',int(uplv)," == ",auto)
    dum2=dat1.loc[:][(dat1.loc[:]['I']==up_as)&(dat1.loc[:]['IP']==lw_as)]
    as_gf=dum2.loc[:]['G*F'].tolist()
    as_A=dum2.loc[:]['A(EK)*SEC'].tolist()
    as_wavelength=dum2.loc[:]['WAVEL/AE'].tolist()
    if len(as_gf)==1: 
        database_updated.at[i,'gf']=as_gf[0]
    else:
        database_updated.at[i,'gf']=0.0
    if len(as_A)==1: 
        database_updated.at[i,'A']=as_A[0]
    else:
        database_updated.at[i,'A']=0.0
    if len(as_wavelength)==1:
        database_updated.at[i,'Wavelength']=as_wavelength[0]
    else
        database_updated.at[i,'Wavelength']=0.0

    print(' ')
```
